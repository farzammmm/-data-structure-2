**نمایش تمام معادلات چندجمله‌ای موجود در لیست‌های پیوندی**


```python
from sympy import sympify

class Node:
    def __init__(self, degree):
        self.degree = degree
        self.coefficients = []
        self.next = None

class PolynomialList:
    def __init__(self):
        self.head = Node(None)
        self.head.next = self.head
        self.n = 0

    def get(self, degree):
        if degree >= self.size():
            raise Exception('Out of list')
        x = self.head.next
        while x.degree != degree:
            x = x.next
        return x

    def insert_after(self, x, coefficients):
        y = Node(len(coefficients) - 1)
        y.coefficients = coefficients
        self.n += 1
        y.next = x.next
        x.next = y
        return y

    def delete(self, x):
        if self.size() == 0:
            raise Exception('List is empty')
        self.n -= 1
        prev_node = self.head
        while prev_node.next != x:
            prev_node = prev_node.next
        prev_node.next = x.next
        return x

    def size(self):
        return self.n

    def is_empty(self):
        return self.n == 0

def get_coefficients(degree):
    while True:
        try:
            expression = input(f"Enter the polynomial for degree {degree} (e.g., 'x^2 + 3x + 2'): ")
            expression = expression.replace("^", "**")  # Convert '^' to '**' for exponentiation
            polynomial = sympify(expression)
            if polynomial.is_polynomial():
                coefficients = [polynomial.coeff(x) for x in reversed(range(degree + 1))]
                return coefficients
            else:
                raise ValueError
        except (ValueError, SyntaxError):
            print("Invalid input! Please enter a valid polynomial expression.")


# Sub menu for inserting a polynomial
def insert_polynomial_submenu():
    print("Insert Polynomial Menu:")
    max_degree = int(input("What is the maximum degree of the polynomial? "))
    polynomial_list = PolynomialList()
    max_coefficients_length = 0
    for i in range(max_degree + 1):
        coefficients = get_coefficients(max_degree - i)
        max_coefficients_length = max(max_coefficients_length, len(coefficients))
        polynomial_list.insert_after(polynomial_list.head, coefficients)
    # Set coefficients length to the maximum length
    for i in range(polynomial_list.size()):
        node = polynomial_list.get(i)
        node.coefficients += [0] * (max_coefficients_length - len(node.coefficients))
    print("Polynomial inserted successfully!")
    return polynomial_list

# Function to sum two polynomials
def sum_polynomials(polynomial_list):
    degree = int(input("Enter the degree of polynomials to be summed: "))
    result = [0] * (degree + 1)
    for i in range(polynomial_list.size()):
        node = polynomial_list.get(i)
        if node.degree <= degree:
            for j in range(len(node.coefficients)):
                result[node.degree - j] += node.coefficients[j]
    print("Result of summing polynomials:")
    print(result)

# Function to multiply two polynomials
def multiply_polynomials(polynomial_list):
  max_degree = int(input("Enter the degree of polynomials to be multiplied: "))
  result = [0] * (2 * max_degree + 1)  # Initialize result list with zeros

  # Iterate over nodes in the polynomial_list
  current_node = polynomial_list.head.next
  while current_node != polynomial_list.head:
    for j in range(len(current_node.coefficients)):
      for k in range(len(current_node.coefficients)):
        degree_sum = current_node.degree + j + k
        if degree_sum < len(result):
          result[degree_sum] += current_node.coefficients[j] * current_node.coefficients[k]
        else:
          result.extend([0] * (degree_sum - len(result) + 1))
    current_node = current_node.next

  print("Result of multiplying polynomials:")
  print(result)



# Function to print all polynomials in the list
def print_polynomials(polynomial_list):
    print("Polynomials in the list:")
    for i in range(polynomial_list.size()):
        node = polynomial_list.get(i)
        print(f"Degree: {node.degree}, Coefficients: {node.coefficients}")

# Main menu
def main_menu():
    print("Main Menu:")
    print("1. Insert a polynomial")
    print("2. Sum two polynomials")
    print("3. Multiply two polynomials")
    print("4. Print")
    print("5. Exit")

# Main function
def main():
    polynomial_list = None
    while True:
        main_menu()
        choice = input("Enter your choice: ")
        if choice == '1':
            polynomial_list = insert_polynomial_submenu()
        elif choice == '2':
            if polynomial_list:
                sum_polynomials(polynomial_list)
            else:
                print("Please insert polynomials first.")
        elif choice == '3':
            if polynomial_list:
                multiply_polynomials(polynomial_list)
            else:
                print("Please insert polynomials first.")
        elif choice == '4':
            if polynomial_list:
                print_polynomials(polynomial_list)
            else:
                print("Polynomial list is empty.")
        elif choice == '5':
            print("Exiting...")
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```
